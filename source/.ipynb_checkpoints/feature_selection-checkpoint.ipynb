```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import time
import pickle
from scipy import stats
import tese_func
from sklearn.feature_selection import mutual_info_classif, f_classif, chi2
import heapq
from tqdm import tqdm
```

```python
data["stage"] = data["stage"].replace(reps)
data["lmr>2.1"] = data["lmr>2.1"].astype('int64')
```

```python
clinic["stage"] = clinic["stage"].replace(reps)
clinic["lmr>2.1"] = clinic["lmr>2.1"].astype('int64')
```

```python
y_clinic = clinic[target]
X_clinic = clinic.drop(columns=[target])
y_gene = gene[target]
X_gene = gene.drop(columns=[target])
X_clinic.head()
```

# Calculate p-values

## p-value of age

```python
x = X_clinic["age"].values.reshape(-1,1)
y = y_clinic.values

mi = mutual_info_classif(x, y, discrete_features=True)[0]
p_val, rv = tese_func.permutation_test(x, y, mi, mutual_info_classif, args={'discrete_features':True},
                             side="right", list_ind=0, rounds=1000)
print("p-value of age: ",p_val)


plt.figure(figsize=(10,7))
ax = plt.hist(rv, bins=20)
plt.axvline(mi, color='k', linestyle='dashed', linewidth=2)
```

## Some gene p-value

```python
x = gene.iloc[:,5].values.reshape(-1,1)
y = gene[target].values
mi = mutual_info_classif(x, y, discrete_features=False)[0]
p_val, rv = tese_func.permutation_test(x, y, mi, mutual_info_classif, args={'discrete_features':False}, rounds=1000)

print("p-value of gene: ", p_val)

plt.figure(figsize=(10,7))
plt.hist(rv, bins=20)
plt.axvline(mi, color='k', linestyle='dashed', linewidth=2)
#min_ylim, max_ylim = plt.ylim()
```

# Calculating p-values of all features
## Mutual Info

```python
y = data[target].values
pvals={}
for col in tqdm(data.drop(columns=[target,"id"]).columns):
    x = data[col].values.reshape(-1,1)
    discrete = x.dtype == np.int64
    mi = mutual_info_classif(x, y, discrete_features=discrete)[0]
    p_val, _ = tese_func.permutation_test(x, y, mi, mutual_info_classif, args={'discrete_features':discrete}, 
                                          side="right", list_ind=0, rounds=500)
    pvals[col]=p_val
    #print("{} : {}".format(col,p_val))


a_file = open("pvalues.pkl", "wb")
pickle.dump(pvals, a_file)
a_file.close()
```

## Wilcoxon Rank Sum Test

#### In statistics, the Mannâ€“Whitney U test (also called the Mannâ€“Whitneyâ€“Wilcoxon (MWW), Wilcoxon rank-sum test, or Wilcoxonâ€“Mannâ€“Whitney test) is a nonparametric test of the null hypothesis that, for randomly selected values X and Y from two populations, the probability of X being greater than Y is equal to the probability of Y being greater than X.
A similar nonparametric test used on dependent samples is the Wilcoxon signed-rank test.



```python
pvals={}
start = time.time()
data0 = data[ data[target]==0]
data1 = data[ data[target]==1]
for col in tqdm(data.drop(columns=[target,"id"]).columns):
    d0 = data0[col]
    d1 = data1[col]
    _, p = stats.mannwhitneyu(d0, d1,alternative="greater")#one sided, right
    pvals[col] = p
    
end = time.time()
b_file = open("pvalues_wil.pkl", "wb")
pickle.dump(pvals, b_file)
b_file.close()
```
