```python
import pandas as pd
import re
import numpy as np
```

```python
ge_data = pd.read_csv("data/GSE132348_series_matrix.txt", sep="\t", header=None, index_col=0).T
print(ge_data.shape)
ge_data.head()
```

```python
file = open("data/GSE132348_series_class.txt", "r")

data = pd.DataFrame()

for l in file.readlines()[1:]:
    l = l.strip('\n').replace('\"', '')
    values = re.split('\t',l)
    if values[0] == "!Sample_characteristics_ch1" or values[0] == "!Sample_geo_accession":
        data[values[1][:3]] = values[1:]

data =  data.drop(columns="sub")
print(data.shape)
data.head()
```

```python
data["gen"] = data["gen"].str.replace("gender: ","")
data["age"] = data["age"].str.replace("age: ","")
data["Sta"] = data["Sta"].str.replace("Stage: ", "")
data["lmr"] = data["lmr"].str.replace("lmr>2.1: ", "")
data["int"] = data["int"].str.replace("interim pet response: ", "")
data.head()
```

```python
data = ge_data.merge(data, right_on="GSM", left_on="ID_REF")
data = data.rename(columns={"ID_REF": "id", "gen": "gender", "Sta": "stage", "lmr":"lmr>2.1", "int":"ipet2"})
data = data.drop(columns="GSM")
print(data.shape)
data.head()
```

```python
data["stage"].value_counts()
```

```python
data = data[ data["ipet2"].isin(["positive", "negative"]) ]         #remove patient without target
data["gender"] = data["gender"].replace({"M":1, "F":0})
data["lmr>2.1"] = data["lmr>2.1"].replace({"yes":1, "no":0, "-":np.nan})
data["ipet2"] = data["ipet2"].replace({"positive":1, "negative":0})
print(data.shape)
data.head()
```

```python
data.to_csv('data/data_ipet.csv', index=False)  
```
