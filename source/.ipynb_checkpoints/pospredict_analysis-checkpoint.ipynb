```python
import pandas as pd
from sklearn.feature_selection import f_classif, mutual_info_classif
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from math import ceil
from IPython.display import Image
import json
from myds import enrichr_func
```

# Loading data

```python
id="id"
data = pd.read_csv("data/data_ipet.csv").set_index(id)
target = "ipet2"

reps = {"I A": 1, "I B":2, "II A":3, "II B":4, "III A":5, "III B":6, "IV A":7, "IV B":8}
selgenes = ["CXCL2", "CXCL3", "CCL18", "TREM1", "SAA1", "PLAU", "FN1",
            "SPP1", "ITGA5", "CD9", "LRP1", "THBS1", "VEGFA"]
palette = ["#4E89AE", "#43658B", "#ED6663", "#FFA372"]

data["stage"] = data["stage"].replace(reps)

print(data.shape)
data.head()
```

### Union10 data

```python
data_union = pd.read_csv("data/dataset_union1.csv").set_index(id)
data_union.head()
```

### Bics 9 1.25 data

```python
data_bics = pd.read_csv("data/bic9_25_euc.csv").set_index(id)
data_bics.head()
```

# Analysis of most discriminative features

### Data Union

```python
df = data_union
n = 12
y = df[target].values
X = df.drop(target, axis=1).values

pval = f_classif(X, y)[1]
tops = pval.argsort()[:n]
tops = np.append(tops, df.columns.get_loc(target))
df = df.iloc[:, tops]

fig, axes = plt.subplots(ceil(n/3), 3, figsize=(20, 16), sharey=True, constrained_layout=True)
axes = axes.flatten()
fig.suptitle(f"Top {n} features", fontsize=16)

for i, c in enumerate(df.columns[:-1]):
    pos = df[ df["ipet2"] == 1 ][c]
    neg =  df[ df["ipet2"] == 0 ][c]
    sns.kdeplot(neg, ax=axes[i], fill=True)
    sns.kdeplot(pos, ax=axes[i], fill=True)
    axes[i].legend(("iPET2 negative", "iPET2 positive"), loc="best")
    axes[i].set_title(c)
    axes[i].set_xlabel("")
plt.savefig("./images/exploratory/12_discriminative_genes.png")
```

## Data Union with mutual info

```python
df = data_union
n = 12
y = df[target].values
X = df.drop(target, axis=1).values

pval = mutual_info_classif(X, y)
tops = pval.argsort()[:n]
tops = np.append(tops, df.columns.get_loc(target))
df = df.iloc[:, tops]

fig, axes = plt.subplots(ceil(n/3), 3, figsize=(20, 16), sharey=True, constrained_layout=True)
axes = axes.flatten()
fig.suptitle(f"Top {n} features", fontsize=16)

for i, c in enumerate(df.columns[:-1]):
    pos = df[ df["ipet2"] == 1 ][c]
    neg =  df[ df["ipet2"] == 0 ][c]
    sns.histplot(neg, ax=axes[i], kde=True)
    sns.histplot(pos, ax=axes[i], kde=True)
    axes[i].legend(("iPET2 negative", "iPET2 positive"), loc="best")
    axes[i].set_title(c)
    axes[i].set_xlabel("")
plt.savefig("./images/exploratory/12_discriminative_genes.png")
```

## In bics 9 1.25 data

```python
with open("./data/bic_patterns/bic9_25_euc.json", "r") as f:
    patterns = json.load(f)
translate = True

df = data_bics
n = 10
y = df[target].values
X = df.drop(target, axis=1).values

pval = f_classif(X, y)[1]
tops = pval.argsort()[:10]
tops = np.append(tops, df.columns.get_loc(target))
df = df.iloc[:, tops]

fig, axes = plt.subplots(ceil(n/3), 3, figsize=(20, 16), sharey=True, constrained_layout=True)
axes = axes.flatten()
fig.suptitle(f"Top {n} features", fontsize=16)

for i, c in enumerate(df.columns[:-1]):
    pos = df[ df["ipet2"] == 1 ][c]
    neg =  df[ df["ipet2"] == 0 ][c]
    sns.kdeplot(neg, ax=axes[i], fill=True)
    sns.kdeplot(pos, ax=axes[i], fill=True)
    axes[i].legend(("iPET2 negative", "iPET2 positive"), loc="best")
    if translate:
        axes[i].set_title(patterns[c], fontsize=16)
    else:
        axes[i].set_title(c)
    axes[i].set_xlabel("")
```

```python
sets = {}
for p in tree_pat:
    sets[str(p)] = patterns[str(p)][0]
titles = list(sets.keys())

lib = 'KEGG_2019_Human'
datasets = enrichr_func.term_graphics(sets, lib, titles=titles, orient="v", return_data= True)
```

```python
d = datasets["1548"]
n_terms = 3
enrichr_func.scatter_genes_term(d, data, target, n_terms, diag_line=True, exp=False, use_axes=True)
enrichr_func.fc_genes_term(d, data, target, n_terms)
enrichr_func.boxplot_genes_term(d, data, target, n_terms)
```
