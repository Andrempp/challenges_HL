```python
NR_COLUMNS: int = 5
HEIGHT: int = 4
def choose_grid(nr):
    if nr < NR_COLUMNS:
        return 1, nr
    else:
        return (nr // NR_COLUMNS, NR_COLUMNS) if nr % NR_COLUMNS == 0 else (nr // NR_COLUMNS + 1, NR_COLUMNS)
```

```python
numeric_vars = gene.iloc[:,1:70].columns
rows, cols = choose_grid(len(numeric_vars))
fig, axs = plt.subplots(rows, cols, figsize=(cols*HEIGHT, rows*HEIGHT))

i, j = 0, 0
for n in range(len(numeric_vars)):
    axs[i, j].set_title('Histogram for %s'%numeric_vars[n])
    axs[i, j].set_xlabel(numeric_vars[n])
    axs[i, j].set_ylabel("nr records")
    axs[i, j].set_ylim(0,50)
    axs[i, j].set_xlim(0,17.5)
    axs[i, j].hist(data[numeric_vars[n]].dropna().values, 'auto')
    i, j = (i + 1, 0) if (n+1) % cols == 0 else (i, j + 1)
plt.show()
```

# Test for normal distribution
### Shapiro-Wilk Test

```python
ps=[]
for i in range(0,len(gene.columns)):
    ps.append(shapiro(gene.iloc[:,i:i+1]).pvalue)

ps = np.array(ps)    
p05 = np.where( ps < 0.05 )[0]
p025 = np.where( ps < 0.025 )[0]
print("percentage less than 0.05 p-value:\t",len(p05)/len(ps) )
print("percentage less than 0.025-value:\t",len(p025)/len(ps) )

plt.figure(figsize=(20,10))
plt.hist(ps, bins=50)
plt.axvline(0.05, color='k', linestyle='dashed', linewidth=2)

plt.show()
```

### If the test is NOT significant, then the data are normal, so any value above .05 indicates normality. If the test is significant (less than .05), then the data are non-normal.

# Fold Change
## All genes

```python
fold_changes = tese_func.logfc_select(gene, target, None)

fig, ax = plt.subplots(figsize=(20,12))
#ax.set_ylim((-0.8,0.8))
sns.barplot(data=pd.DataFrame(fold_changes).T,  palette="Blues_d", ax=ax)
ax.set(xlabel='Genes', ylabel='log( FoldChange )')
```

```python
fig = plt.figure(figsize=(20,12))

fc_genes = tese_func.logfc_select(gene, target, threshold=2)
fc_genes = fc_genes.reindex(fc_genes.abs().sort_values().index) #order by absolute
ax = sns.barplot(data=pd.DataFrame(fc_genes).T,  palette="Blues_d")
ax.set(xlabel='Genes', ylabel='log( FoldChange )')
```

```python
def fold_change(x, y, log=False, constant=0.001):
    neg_ind = np.where(y==0)
    pos_ind = np.where(y==1)
    if log:
        neg = np.log2(x[neg_ind] + constant).mean()
        pos = np.log2(x[pos_ind] + constant).mean()
    else:
        neg = x[neg_ind].mean()
        pos = x[pos_ind].mean()
    res = pos - neg
    return res

y = gene[target].values

fcs = []
p_vals = []
for g in gene.drop(columns=[target]).columns:
    x = gene[g].values.reshape(-1,1)
    fc = fold_change(x, y)
    p_val, _ = tese_func.permutation_test(x, y, fc, fold_change, rounds=1000)
    fcs.append(fc)
    p_vals.append(p_val)


volc = pd.DataFrame()
volc["gene_symbol"] = gene.drop(columns=[target]).columns
volc["log2FoldChange"] = fcs
volc["pvalue"] = p_vals
volc.to_csv("volcano.csv", index=False)
```
